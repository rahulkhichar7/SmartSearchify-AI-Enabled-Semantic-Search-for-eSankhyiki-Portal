```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from joblib import load
import warnings
import ast

# Suppress warnings
warnings.filterwarnings('ignore')

# Load your dataframe (replace with your actual data loading method)
# df = pd.read_csv('your_data.csv')  # Update with your actual data source
# df['search_emb'] = df['search_emb'].apply(ast.literal_eval)  # Convert string to list

# Load the pre-trained classifier
classifier = load('LinearSVC_classifier.joblib')

# Initialize the sentence transformer model
embedder = SentenceTransformer("all-MiniLM-L6-v2")

# Unique products (classes)
products = np.array(['ASI', 'ASUSE', 'CAMS', 'CPI', 'HCES', 'IIP', 'MIS', 'NAS', 'PLFS', 'WMI'])

def classify_query(query):
    # Embed the query
    query_embedding = embedder.encode([query])
    
    # Predict probabilities (using decision function for LinearSVC)
    decision_scores = classifier.decision_function(query_embedding)
    
    # Convert decision scores to probabilities (softmax)
    exp_scores = np.exp(decision_scores - np.max(decision_scores))
    probabilities = exp_scores / exp_scores.sum(axis=1, keepdims=True)
    probabilities = probabilities[0]  # Get the first (and only) row
    
    # Get top 3 predictions
    top3_indices = np.argsort(probabilities)[-3:][::-1]
    top3_products = products[top3_indices]
    top3_probs = probabilities[top3_indices]
    
    return list(zip(top3_products, top3_probs))

def semantic_search(query_embedding, product_df, n_results):
    # Convert embeddings to numpy array
    doc_embeddings = np.array(product_df['search_emb'].tolist())
    
    # Calculate cosine similarities
    similarities = cosine_similarity(query_embedding, doc_embeddings)[0]
    
    # Get top n results
    top_indices = np.argsort(similarities)[-n_results:][::-1]
    results = product_df.iloc[top_indices]['Title'].tolist()
    
    return results

def get_relevant_results(query, top_products):
    query_embedding = embedder.encode([query])
    results = {}
    
    # Determine how many results to return for each product
    n_results = [3, 2, 1]  # For 1st, 2nd, 3rd product respectively
    
    for i, (product, prob) in enumerate(top_products):
        product_df = df[df['Product'] == product]
        num_rows = len(product_df)
        
        if num_rows <= 50:
            # Directly apply semantic search on all rows
            product_results = semantic_search(query_embedding, product_df, n_results[i])
        else:
            # First filter to top 50 most similar embeddings
            doc_embeddings = np.array(product_df['search_emb'].tolist())
            similarities = cosine_similarity(query_embedding, doc_embeddings)[0]
            top50_indices = np.argsort(similarities)[-50:][::-1]
            top50_df = product_df.iloc[top50_indices]
            
            # Then apply semantic search on these 50
            product_results = semantic_search(query_embedding, top50_df, n_results[i])
        
        results[product] = {
            'probability': prob,
            'results': product_results
        }
    
    return results

def main():
    query = input("Enter your query: ")
    
    # Step 1: Classify the query
    top_products = classify_query(query)
    print("\nTop 3 predicted products with probabilities:")
    for product, prob in top_products:
        print(f"{product}: {prob:.4f} ({(prob*100):.1f}%)")
    
    # Step 2: Get relevant results for each product
    relevant_results = get_relevant_results(query, top_products)
    
    # Step 3: Display results
    print("\nRelevant results:")
    for product, data in relevant_results.items():
        print(f"\n{product} (probability: {(data['probability']*100):.1f}%):")
        for i, result in enumerate(data['results'], 1):
            print(f"  {i}. {result}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import ast

embedder = SentenceTransformer("all-MiniLM-L6-v2")

def search_and_select(query):
    # Embed the query
    query_embedding = embedder.encode([query])
    
    # Convert all document embeddings to numpy array
    doc_embeddings = np.array(df['search_emb'].tolist())
    
    # Calculate cosine similarities
    similarities = cosine_similarity(query_embedding, doc_embeddings)[0]
    
    # Get top 50 most similar documents
    top50_indices = np.argsort(similarities)[-50:][::-1]
    top50_df = df.iloc[top50_indices].copy()
    top50_df['similarity'] = similarities[top50_indices]
    
    # Get top 4 most similar titles from these 50
    top4_indices = np.argsort(top50_df['similarity'])[-4:][::-1]
    top4_results = top50_df.iloc[top4_indices]
    
    # Display the top 4 results
    print("\nTop 4 most relevant results:")
    for i, (idx, row) in enumerate(top4_results.iterrows(), 1):
        print(f"\n{i}. {row['Title']}")
        print(f"   Similarity: {row['similarity']:.4f}")
        print(f"   Product: {row['Product']}")
        print(f"   Category: {row['Category']}")
    
    # Let user select one
    while True:
        try:
            selection = int(input("\nEnter the number of the result you want to see details (1-4), or 0 to exit: "))
            if selection == 0:
                return
            elif 1 <= selection <= 4:
                selected_row = top4_results.iloc[selection-1]
                display_full_details(selected_row)
                break
            else:
                print("Please enter a number between 1 and 4, or 0 to exit.")
        except ValueError:
            print("Please enter a valid number.")

def display_full_details(row):
    print("\nFull details of selected result:")
    print("-" * 50)
    print(f"Title: {row['Title']}")
    print(f"Product: {row['Product']}")
    print(f"Category: {row['Category']}")
    print(f"Geography: {row['Geography']}")
    print(f"Frequency: {row['Frequency']}")
    print(f"Reference Period: {row['Reference Period']}")
    print(f"Release Date: {row['Release Date']}")
    print(f"Table No: {row['Table No']}")
    print(f"Download URL: {row['Download URL']}")
    print(f"Data Source: {row['Data Source']}")
    print(f"Description: {row['Description']}")
    print("-" * 50)

# Main program
if __name__ == "__main__":
    print("Semantic Search System")
    print("---------------------")
    while True:
        query = input("\nEnter your search query (or 'exit' to quit): ")
        if query.lower() == 'exit':
            break
        search_and_select(query)
```
